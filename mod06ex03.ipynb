```python
import pandas as pd
import requests
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[(sinasc['APGAR5'] <= 3), 'CONDICAO'] = 'ASFIXIA GRAVE'
sinasc.loc[(sinasc['APGAR5'] >  3), 'CONDICAO'] = 'ASFIXIA MODERADA'
sinasc.loc[(sinasc['APGAR5'] >  5), 'CONDICAO'] = 'ASFIXIA LEVE'
sinasc.loc[(sinasc['APGAR5'] >  7), 'CONDICAO'] = 'NORMAL'

sinasc
```

```python
sinasc['CONDICAO']
```

```python
sns.countplot(x='CONDICAO', data=sinasc)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# histograma com covariáveis
sns.scatterplot(data = sinasc
            ,x = 'CONDICAO', y = 'GESTACAO'
            ,hue = 'SEXO'
            ,alpha = 0.2
           ).tick_params(axis="x", rotation = 90)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Tabela cruzada
tab_cruz = pd.crosstab(sinasc['CONDICAO'], sinasc['GESTACAO'])
tab_cruz
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tab_cruz.sum(axis = 1)
```

```python
tab_cruz_pct = tab_cruz.div(tab_cruz.sum(axis = 1), axis = 0)
tab_cruz_pct
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tab_cruz_pct.plot.bar()
# concordo, é visivel o maior indice de asfixias nas gestacoes anteriores a 37 semanas.
```
