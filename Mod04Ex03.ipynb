````python
#1) Seu código aqui
def consulta_tipo(a,b):
  return df_pvl_todos[df_pvl_todos['tipo_interessado'] == a][df_pvl_todos['uf'] == b]

consulta_tipo('Município','GO')
````

````python
# 2) Seu código aqui
df_estado_mg = consulta_tipo('Estado','MG')
df_estado_mg['status'].value_counts()
````
