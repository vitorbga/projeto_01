1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

# Fixing random state for reproducibility
np.random.seed(2360873)
#dataframe de previsao de renda
prev_renda = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
prev_renda.head()
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=prev_renda, dodge=True, ci = 90)
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=prev_renda, dodge=True, ci = 90)
````

````python
ax = sns.pointplot(x="tipo_renda", y="renda", hue = 'posse_de_veiculo',
                   data=prev_renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
ax = sns.pointplot(x="tipo_renda", y="renda", hue = 'posse_de_imovel',
                   data=prev_renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

De acordo coms os gráficos criados das variáveis qualitativas (posse de imóvel e posse de veículo), discrimina-se renda de forma estatisticamente significante.

Nota-se que, a diferença de renda de posse de veículo é maior que em posse de ímovel. Assim, a população deste exemplo opta por adquirir um ímovel antes de um veículo.
